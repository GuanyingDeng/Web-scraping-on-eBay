```python
x = []
z = []
for pgn in range(1, 11): 
    url = "https://www.ebay.com/sch/i.html?_nkw=canon+eos+5d&LH_BIN=1&_ipg=100"+ '&_pgn='+str(pgn)
    user_agent = {'User-agent': 'Mozilla/5.0'} 
    response = requests.get(url, headers = user_agent)
    soup = BeautifulSoup(response.content)
    items = soup.select("#srp-river-results > ul > li.s-item")
    for item in items:
        a = item.find("a")
        text = item.find("span").get_text()
        if re.search('.*S.*P.*O.*N.*S.*O.*R.*E.*D.*',text): 
        #print(a['href']) 
            url_simplify = re.sub("(.*)\\?.*", r"\1", a['href'])
            x.append(url_simplify)
        else:
            url_simplify_nonspon = re.sub("(.*)\\?.*", r"\1", a['href'])
            z.append(url_simplify_nonspon)
with open("sponsored.txt", 'w') as file:
    for item in x:
        file.write('%s\n' % item)
        
with open("non-sponsored.txt", 'w') as file:
    for item in z:
        file.write('%s\n' % item)
```

```python
##d
from bs4 import BeautifulSoup
path='sponsored'
files=os.listdir(path)
files=[name for name in files if name != ".DS_Store"]
rows=[]
for file in files:
    try:
        ##get id of the items
        item_id=file.split(".")[0]
        item_id=int(item_id)
        
        file=open(os.getcwd()+'/sponsored/'+file,'r')
        e=file.read()
        soup = BeautifulSoup(e, 'html.parser')
        
        #seller name
        seller_name=soup.find("span", attrs= {"class": "mbg-nw"})
        if seller_name is None:
            seller_name=soup.find_all("span", attrs= {"class": "no-wrap"})[1].text
            lst=seller_name.split('(')
            seller_name=lst[0]
        else:
            seller_name=seller_name.text
        
        #seller score
        score =soup.find("span", attrs= {"class": "mbg-l"})
        if score is None:
            score=soup.find_all("span", attrs= {"class": "no-wrap"})[1].text
            lst=score.split('(')
            score=lst[1]
            lst=score.split(')')
            score=lst[0]
        else:
            score=score.text
            score=score.split("(")[1]
            score=score.split(")")[0]
            score=re.search('\d+',score)[0]
        score=int(score)
        
        #item price
        price=soup.find("span",attrs={"id": "prcIsum"})
        if price is None:
            price=soup.find("span",attrs={"id": "mm-saleDscPrc"})
        if price is None:
            price=soup.find("h2",attrs={"class": "display-price"})
        if price is None:
            price=soup.find("span",attrs={"id": "prcIsum_bidPrice"})
        price=price.text
        if "$" in price:
            if " " in  price:
                price=price.split("$")[1]
            else:
                price=price[1:]
        else:
            price=price.split(' ')[1]
        if "," in price:
            price=price.split(',')
            price=''.join(price)
        if 'ea' in price:
            price=price.split('/')
            price=price[0]
        price=float(price)
        price=price*100
        price=int(price)

        # #items sold
        s=soup.find("a",attrs={"class": "vi-txt-underline"})
        if s is None:
            num_sold=None
        else:
            sold = s.text
            num_sold = sold.split(" ")[0]
            if ',' in num_sold:
                num_sold=num_sold.split(",")
                num_sold=''.join(num_sold)
            num_sold = int(num_sold)
        
        #best offer available
        bo=soup.find("div",attrs={"class": "vi-bbox-dspn u-flL lable boLable"})
        if bo is None:
            best_offer="NO"
        else:
            best_offer="YES"
        
        #title
        title=soup.find("h1",attrs={"id":"itemTitle"})
        if title is None:
            title=soup.find("h1",attrs={"class":"product-title"})
            title=title.text
        else:
            title=title.text
            title=title.split("\xa0")[1]
        
        #returns allowed
        return_allowed=soup.find("span",attrs={"id": "vi-ret-accrd-txt"})
        if return_allowed is None:
            return_allowed=None
        else:
            return_allowed=return_allowed.text
#             return_list=return_allowed.split('\xa0')
#             print(return_list)
#             return_allowed=return_list.join(' ')
        if return_allowed is "None":
            return_allowed=return_allowed
        elif return_allowed in ('Free 30 day returns','30 day\xa0returns.\xa0Buyer pays for return shipping','14 day\xa0returns.\xa0Buyer pays for return shipping','Free 60 day returns','60 day\xa0returns.\xa0Buyer pays for return shipping'):
            return_allowed="YES"
        else:
            return_allowed="NO"
        
        #shipping price
        ship=soup.find("span",attrs={"id":"fshippingCost"})
        if ship is None:
            shipping="None"
        else:
            shipping=ship.text
            if shipping=='\nFREE\n':
                shipping='FREE'
            else:
                shipping=re.search('\d+.\d+',shipping)[0]
                shipping=float(shipping)
        
        #condition
        cond=soup.find("div",attrs={"id": "vi-itm-cond"})
        if cond is None:
            cond="New"
        else:
            cond=cond.text
        
        row=[item_id,seller_name,score,price,num_sold,best_offer,title,return_allowed,shipping,cond,"sponsored"]
        rows.append(row)
        
    except:
        print("error")
        print(file)
        continue
```

```python
##f
##return_allowed
cursor = connection.cursor()
cursor.execute("select count(a.return_allowed_binary), avg(a.return_allowed_binary), min(a.return_allowed_binary), max(a.return_allowed_binary), a.sponsored, a.item_condition \
                from (select (case when return_allowed ='YES' then 1 when return_allowed ='NO' then 0 end) return_allowed_binary,sponsored, item_condition \
                from eBay.eBay_items) a group by a.sponsored, a.item_condition;")
result=cursor.fetchall()
result_df=pd.DataFrame(result)
result_df.columns=['count_whether_return_allowed', 'mean_whether_return_allowed', 'min_whether_return_allowed','max_whether_return_allowed', 'whether_sponsored', 'condition']
result_df
```

```python
##best_offer_available
cursor = connection.cursor()
cursor.execute("select count(b.best_offer_binary), avg(b.best_offer_binary), min(b.best_offer_binary), max(b.best_offer_binary), b.sponsored, b.item_condition \
            from (select (case when best_offer ='YES' then 1 when best_offer ='NO' then 0 end) best_offer_binary,sponsored, item_condition \
            from eBay.eBay_items) b group by b.sponsored, b.item_condition;")
result_bo=cursor.fetchall()
bo_df=pd.DataFrame(result_bo)
bo_df.columns=['count_whether_best_offer', 'mean_whether_best_offer', 'min_whether_best_offer','max_whether_best_offer', 'whether_sponsored', 'condition']
bo_df
```
